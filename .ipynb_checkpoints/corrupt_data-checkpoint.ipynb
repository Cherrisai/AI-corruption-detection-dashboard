```python
import pandas as pd
import numpy as np
import random

states = {

"Tamil Nadu":["Chennai","Coimbatore","Madurai","Trichy"],
"Karnataka":["Bangalore","Mysore","Mangalore","Hubli"],
"Maharashtra":["Mumbai","Pune","Nagpur","Nashik"],
"Gujarat":["Ahmedabad","Surat","Vadodara","Rajkot"],
"Rajasthan":["Jaipur","Udaipur","Jodhpur","Kota"],
"Uttar Pradesh":["Lucknow","Kanpur","Agra","Varanasi"],
"Telangana":["Hyderabad","Warangal","Nizamabad","Karimnagar"],
"West Bengal":["Kolkata","Howrah","Durgapur","Siliguri"],
"Kerala":["Kochi","Trivandrum","Kozhikode","Thrissur"],
"Delhi":["New Delhi","Dwarka","Rohini","Saket"]

}

categories = ["Road","Bridge","Government Building","Park"]

years = [2020,2021,2022,2023,2024,2025,2026]

rows=[]
project_id=1

for year in years:

    for state in states:

        for city in states[state]:

            for category in categories:

                for i in range(80):

                    # ROAD DATA
                    road_length = round(np.random.uniform(2,40),2)
                    road_lanes = random.choice([2,4,6])
                    road_type = random.choice(["Highway","City Road","Rural Road"])

                    # BRIDGE DATA
                    bridge_length = round(np.random.uniform(200,2000),2)
                    bridge_type = random.choice([
                        "Flyover",
                        "River Bridge",
                        "Railway Overbridge"
                    ])

                    # BUILDING DATA
                    building_area = round(np.random.uniform(15000,90000),2)
                    floors = random.randint(2,10)
                    building_type = random.choice([
                        "Hospital",
                        "School",
                        "Office",
                        "Court"
                    ])

                    # PARK DATA
                    park_area = round(np.random.uniform(3,50),2)
                    park_facilities = random.choice([
                        "Playground",
                        "Lake",
                        "Walking Track",
                        "Garden",
                        "Open Gym"
                    ])

                    # COST DATA
                    material_cost = np.random.randint(3000000,80000000)
                    labour_cost = np.random.randint(800000,35000000)
                    equipment_cost = np.random.randint(500000,12000000)

                    expected_cost = material_cost + labour_cost + equipment_cost

                    corruption_factor = np.random.uniform(1.1,1.6)

                    allocated_budget = int(expected_cost * corruption_factor)

                    actual_cost = int(expected_cost * np.random.uniform(0.9,1.05))

                    corruption_amount = allocated_budget - actual_cost

                    corruption_score = round(corruption_amount / allocated_budget,3)

                    population = np.random.randint(200000,15000000)

                    traffic_density = np.random.randint(10,100)

                    rows.append([
                        project_id,
                        state,
                        city,
                        year,
                        category,

                        road_length,
                        road_lanes,
                        road_type,

                        bridge_length,
                        bridge_type,

                        building_area,
                        floors,
                        building_type,

                        park_area,
                        park_facilities,

                        material_cost,
                        labour_cost,
                        equipment_cost,

                        expected_cost,
                        allocated_budget,
                        actual_cost,

                        corruption_amount,
                        corruption_score,

                        population,
                        traffic_density
                    ])

                    project_id += 1


columns=[

"project_id",
"state",
"city",
"year",
"project_category",

"road_length_km",
"road_lanes",
"road_type",

"bridge_length_m",
"bridge_type",

"building_area_sqft",
"floors",
"building_type",

"park_area_acres",
"park_facilities",

"material_cost_rs",
"labour_cost_rs",
"equipment_cost_rs",

"expected_cost_rs",
"allocated_budget_rs",
"actual_cost_rs",

"corruption_amount_rs",
"corruption_risk_score",

"population",
"traffic_density"

]

df=pd.DataFrame(rows,columns=columns)

df.to_csv("india_clean_projects_2020_2026.csv",index=False)

print("Dataset created successfully")
print("Total Rows:",len(df))
```
